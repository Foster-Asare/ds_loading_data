## Indexing / Slicing / selecting sections of the dataframe
- Access a single value from a row/column pair (.at)
- Access a single value from a row/column pair by integer positionn (.iat)
- Access a group of rows and columns by label (s) or a boolean array. (.loc)

```python
## Getting a single value by the row number and column name
toyota_df.at[4, 'FuelType']
```

```python
toyota_df.at[9, 'KM']
```

```python
toyota_df.iat[4, 4]
```

```python
## all the values in the FuelType column
toyota_df.loc[15:30, 'FuelType']
```
